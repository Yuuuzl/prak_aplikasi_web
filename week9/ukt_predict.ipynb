```python
# Importing libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
data = pd.read_csv('ukt_dataset.csv')
data

```

```python
# Split the data into features and target
X = data[['pendapatan_orangtua']]
y = data[['ukt']]

```

```python
# Plot Pairplot (Variables Relationship)
sns.pairplot(data)
```

```python
# Train the model
model = LinearRegression()
model.fit(X, y)

```

```python
# Correlation Matrix
corr_matrix = data.corr()
sns.heatmap(corr_matrix, annot = True)
plt.show()
```

```python
# Save the trained model
joblib.dump(model, 'linear_regression_ukt.pkl')
```
